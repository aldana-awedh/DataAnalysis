```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

df = pd.read_csv(r"C:\Users\aldan\OneDrive\Desktop\Data Analyst Bootcamp\Python Files\world_population.csv")
```

```python
# Remove scientific notation 
pd.set_option('display.float_format', lambda x: '%.2f' %x)
```

```python
# Display general information about dataset
df.info()
```

```python
# Display statistical information about dataset
df.describe()
```

```python
# Display number of null values per column in the dataset
df.isnull().sum()
```

```python
# Display number of unique values per column in the dataset
df.nunique()
```

```python
# Extract top 10 entries by "2022 Population"
df.sort_values(by='2022 Population', ascending=False).head(10)
```

```python
# Display correlation between all numeric columns in the dataset
df.corr(numeric_only=True)
```

```python
# Create heatmap to visualize correlation between numeric columns
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Group data by "Continent" and display the average values of each numeric column per continent 
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
# Filter dataset for "Oceania" continent 
df[df['Continent'].str.contains('Oceania')]
```

```python
# Visualize population trend per "Continent" 
df_grouped = df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)[reversed(df.columns[5:13])]
df_transposed = df_grouped.transpose()
df_transposed.plot(figsize = (15,7))
```

```python
# Visualize all numeric data in boxplots 
df.boxplot(figsize = (20,10))
```
