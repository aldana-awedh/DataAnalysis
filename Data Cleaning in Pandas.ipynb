```python
import pandas as pd 
```

```python
df = pd.read_excel(r"C:\Users\aldan\OneDrive\Desktop\Data Analyst Bootcamp\Python Files\Customer Call List.xlsx")
df
```

```python
# Drop duplicates 
df.drop_duplicates(inplace = True)
```

```python
# Drop unnecessary columns 
df.drop('Not_Useful_Column', axis = 1, inplace = True)
```

```python
# Remove numeric values and special characters "Last_Name" column 
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
# Remove non-alphanumeric characters from "Phone_Number" columns 
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex = True)
```

```python
# Convert "Phone_Number" column to string format 
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
# Format numbers as "123-456-7890"
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# Remove values that are not phone numbers 
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
```

```python
# Split "Address" column into separate columns 
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand = True)
```

```python
# Standardize "Paying Customer" column 
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y').str.replace('No','N')
```

```python
# Standardize "Do_Not_Contact" column 
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y').str.replace('No','N')
```

```python
# Remove all invalid values 
df.replace('N/a', '', inplace = True)
df.fillna('', inplace = True)
```

```python
# Remove all entries for customers who did not provide a phone number or cannot be contacted 
df = df[(df['Do_Not_Contact'] != 'Y') & (df['Phone_Number'] != '')]
```

```python
# Drop "Address" column 
df.drop('Address', axis = 1, inplace = True)
```

```python
# Reset index 
df.reset_index(inplace =  True, drop = True)
```
